```python
# 1. Importar librer√≠as
import pandas as pd
import numpy as np

# 2. Cargar dataset base
df = pd.read_csv("../data/processed/dataset_part_2.csv")
print("Dimensiones iniciales:", df.shape)
df.head()
```

```python
# 3. Seleccionar columnas relevantes
cols = [
    "FlightNumber", "PayloadMass", "Orbit", "LaunchSite",
    "GridFins", "Reused", "Legs", "Class"
]
df = df[cols]
print("Columnas seleccionadas:", df.columns.tolist())
```

```python
# 4. Renombrar columnas a nombres claros
df = df.rename(columns={
    "FlightNumber": "flight",
    "PayloadMass": "payload",
    "Orbit": "orbit",
    "LaunchSite": "site",
    "GridFins": "gridfins",
    "Reused": "reused",
    "Legs": "legs",
    "Class": "success"
})
df.head()
```

```python
# 5. Manejo de valores nulos
df["payload"] = df["payload"].fillna(df["payload"].median())
df[["gridfins", "reused", "legs"]] = df[["gridfins", "reused", "legs"]].fillna(False).astype(bool)
df["orbit"] = df["orbit"].fillna("Unknown").replace("", "Unknown")
df["site"]  = df["site"].fillna("Unknown").replace("", "Unknown")

# ‚úÖ Verificaci√≥n de limpieza
print("Valores √∫nicos en orbit:", df["orbit"].unique())
print("N√∫mero de NaN en orbit:", df["orbit"].isna().sum())
print("Filas con NaN en orbit:\n", df[df["orbit"].isna()])
# ‚úÖ Verificaci√≥n de limpieza en SITE
print("Valores √∫nicos en site:", df["site"].unique())
print("N√∫mero de NaN en site:", df["site"].isna().sum())
print("Filas con NaN en site:\n", df[df["site"].isna()])

```

```python
# 6. Verificaci√≥n final
df.info()
df.head()
```

```python
# 7. Guardar dataset limpio
output_path = "../data/processed/clean_dataset.csv"
df.to_csv(output_path, index=False)
print(f"‚úÖ Dataset limpio guardado en {output_path}")
```
